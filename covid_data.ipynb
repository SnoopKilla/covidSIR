```python
import numpy as np
import matplotlib.pyplot as plt
from src.sampler import mcmc_sampler
```

## Germany Data

```python
# Import data.
P = 83000000
i = np.genfromtxt("data/germany_infected.csv", dtype=np.int32)
r = np.genfromtxt("data/germany_removed.csv", dtype=np.int32)
s = np.repeat(P, i.shape[0]) - i - r
```

### One Breakpoint

```python
# Define parameters.
d = 2
sigma = 0.01
alpha = np.repeat(2, d)
beta = np.repeat(0.1, d)
a = 1
b = 1
M = 3
phi = 0.995

# Run MCMC.
n_iterations = 50000
burnin = 10000
p, lam, t, lam_ar, t_ar = mcmc_sampler(s, i, d, P, n_iterations, burnin, M, sigma, alpha, beta, a, b, phi)

# Print the results.
print(f"Acceptance rate for lambda: {lam_ar}")
print(f"Acceptance rate for t: {t_ar}")
print(f"Estimated lambda: {np.average(lam,axis = 1)}")
print(f"Estimated t: {np.average(t,axis = 1)}")
print(f"Estimated p: {np.average(p)}")

# Plot the series.
fig, axs = plt.subplots(nrows=2)
fig.set_figheight(15)
fig.set_figwidth(15)
ax1_left = axs[0]
ax2_left = axs[1]
ax1_right = ax1_left.twinx()
ax2_right = ax2_left.twinx()
ax1_left.plot(s, color='red', label="Susceptible")
ax1_right.plot(i, color='blue', label="Infected")
ax1_left.legend(loc=2)
ax1_right.legend(loc=1)
delta_i = -np.diff(s)
ax2_left.plot(delta_i, color="blue", label="Newly Infected Individuals")
ax2_right.plot(i, color='blue', linestyle='dashed', label="Infected")
ax2_left.legend(loc=2)
ax2_right.legend(loc=1)

# Display obtained breakpoints on plot.
for breakpoint in np.average(t, axis=1):
  ax1_right.axvline(breakpoint, color="green")
  ax2_right.axvline(breakpoint, color="green")
plt.show()
```

For the Germany data, the model with 3 breakpoints seems to fit the data better. Indeed, we know that in Germany:
- Around 9 March 2020, large public events (as soccer matches) were canceled;
- Around 16 March 2020, schools, childcare facilities and many stores were closed;
- On 23 March 2020, a far-reaching contact ban was imposed by government authorities.
